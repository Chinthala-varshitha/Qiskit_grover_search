```python
import math
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

# Qiskit 2.x imports
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error, ReadoutError

# ---------------------------
# Utilities
# ---------------------------
def plot_counts(counts, title="Counts", top_n=12):
    if not counts:
        print("No counts to plot.")
        return
    items = sorted(counts.items(), key=lambda kv: kv[1], reverse=True)[:top_n]
    labels = [k for k, _ in items]
    vals = [v for _, v in items]
    plt.figure(figsize=(max(6, 0.6 * len(labels)), 3.5))
    plt.bar(labels, vals)
    plt.title(title)
    plt.xlabel("bitstring (classical)")
    plt.ylabel("counts")
    plt.show()

def most_likely(counts):
    if not counts:
        return None
    return max(counts.items(), key=lambda kv: kv[1])[0]

def matches_target(key, target):
    """Return True if measured key matches target in either orientation (Qiskit bit-order caveat)."""
    return key == target or key[::-1] == target

# ---------------------------
# Grover Oracle & Diffuser
# ---------------------------
def mark_state_oracle(qc, qr, marked_bitstring):
    n = len(qr)
    if len(marked_bitstring) != n:
        raise ValueError("marked_bitstring length must equal number of qubits")
    # Apply X on qubits where target bit == '0' to map the marked state to |11...1>
    for i, bit in enumerate(marked_bitstring):
        if bit == '0':
            qc.x(qr[i])
    # Multi-controlled Z using H on target to convert MCX to MCZ
    if n == 1:
        qc.z(qr[0])
    else:
        qc.h(qr[-1])
        qc.mcx(qr[:-1], qr[-1])  # controls: all but last; target: last
        qc.h(qr[-1])
    # Undo the Xs
    for i, bit in enumerate(marked_bitstring):
        if bit == '0':
            qc.x(qr[i])

def diffuser(qc, qr):
    n = len(qr)
    # H on all
    qc.h(qr)
    # X on all
    qc.x(qr)
    # multi-controlled Z on all qubits
    if n == 1:
        qc.z(qr[0])
    else:
        qc.h(qr[-1])
        qc.mcx(qr[:-1], qr[-1])
        qc.h(qr[-1])
    # Undo X and H
    qc.x(qr)
    qc.h(qr)

# ---------------------------
# Build Grover circuit
# ---------------------------
def grover_circuit(n, marked, iterations=1, measure=True, draw_gate=False):
    qr = QuantumRegister(n, 'q')
    cr = ClassicalRegister(n, 'c') if measure else None
    qc = QuantumCircuit(qr, cr) if measure else QuantumCircuit(qr)

    # Initialize uniform superposition
    qc.h(qr)

    # Grover iterations
    for _ in range(iterations):
        mark_state_oracle(qc, qr, marked)
        diffuser(qc, qr)

    if measure:
        qc.measure(qr, cr)

    if draw_gate:
        # qc.draw('mpl') returns a matplotlib Figure if backend supports it
        fig = qc.draw('mpl', fold=120)
        return qc, fig
    return qc

# ---------------------------
# Simulation runners
# ---------------------------
def run_ideal(qc, shots=1024):
    backend = AerSimulator()
    tc = transpile(qc, backend=backend)
    job = backend.run(tc, shots=shots)
    res = job.result()
    return res.get_counts()

def make_simple_noise_model(p1=0.002, p2=0.01, ro=0.03):
    nm = NoiseModel()
    single = depolarizing_error(p1, 1)
    two = depolarizing_error(p2, 2)
    nm.add_all_qubit_quantum_error(single, ['u3','u2','u1','h','x'])
    nm.add_all_qubit_quantum_error(two, ['cx'])
    nm.add_all_qubit_readout_error(ReadoutError([[1-ro, ro], [ro, 1-ro]]))
    return nm

def run_noisy(qc, noise_model, shots=2048):
    backend = AerSimulator()
    tc = transpile(qc, backend=backend)
    job = backend.run(tc, shots=shots, noise_model=noise_model)
    res = job.result()
    return res.get_counts()

# ---------------------------
# Experiment runner / examples
# ---------------------------
def experiment(n, marked_states, iterations_list=None, plot=True, noisy=False):
    N = 2**n
    if iterations_list is None:
        k_opt = max(1, int(math.floor(math.pi / 4 * math.sqrt(N))))
        iterations_list = [k_opt, max(1, k_opt-1), k_opt+1]  # try around theoretical optimum

    noise_model = make_simple_noise_model() if noisy else None

    results = {}
    for marked in marked_states:
        if len(marked) != n:
            raise ValueError(f"Marked state '{marked}' length != n ({n})")
        results[marked] = {}
        for k in iterations_list:
            print(f"\n=== n={n}, marked={marked}, iterations={k} ===")
            qc, fig = grover_circuit(n, marked, iterations=k, measure=True, draw_gate=True)
            # Draw circuit (mpl)
            print("Circuit visualization (mpl figure shown).")
            display(fig)  # works in Jupyter; in scripts, the fig will appear if plt.show() is used
            # Ideal run
            counts_ideal = run_ideal(qc, shots=2048)
            print("Ideal counts (top 8):", dict(list(sorted(counts_ideal.items(), key=lambda kv: kv[1], reverse=True))[:8]))
            if plot:
                plot_counts(counts_ideal, title=f"Ideal: n={n}, marked={marked}, k={k}")
            results[marked][(k, 'ideal')] = counts_ideal

            if noisy:
                counts_noisy = run_noisy(qc, noise_model, shots=4096)
                print("Noisy counts (top 8):", dict(list(sorted(counts_noisy.items(), key=lambda kv: kv[1], reverse=True))[:8]))
                if plot:
                    plot_counts(counts_noisy, title=f"Noisy: n={n}, marked={marked}, k={k}")
                results[marked][(k, 'noisy')] = counts_noisy

            # Print success prob (check both orientations of bitstring)
            ml_ideal = most_likely(counts_ideal)
            success_ideal = sum(v for k2, v in counts_ideal.items() if matches_target(k2, marked)) / sum(counts_ideal.values())
            print(f"Ideal success probability for marked='{marked}' = {success_ideal:.4f} (most likely='{ml_ideal}')")

            if noisy:
                ml_noisy = most_likely(counts_noisy)
                success_noisy = sum(v for k2, v in counts_noisy.items() if matches_target(k2, marked)) / sum(counts_noisy.values())
                print(f"Noisy success probability for marked='{marked}' = {success_noisy:.4f} (most likely='{ml_noisy}')")
    return results

# ---------------------------
# Example usage
# ---------------------------
if __name__ == "__main__":
    # 1) For 3 qubits, try different marked states
    marked_states_3 = ['001', '111', '010']
    res3 = experiment(n=3, marked_states=marked_states_3, iterations_list=[1,2], plot=True, noisy=True)

    # 2) Extend to 4 qubits (update oracle & diffuser work for generic n)
    # Recommended iterations ~ floor(pi/4 * sqrt(2^n))
    marked_states_4 = ['0011', '1111']  # example marked states
    # compute recommended k for n=4: floor(pi/4 * sqrt(16)) = floor(pi/4 * 4) = floor(pi) = 3
    res4 = experiment(n=4, marked_states=marked_states_4, iterations_list=[1,2,3,4], plot=True, noisy=True)

    print("\nDone experiments. Inspect `res3` and `res4` dicts for raw counts.")
```
